```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#calculations
Number_of_Unique_Items = len(purchase_data["Item ID"].unique())
Average_Purchase_Price = purchase_data["Price"].mean()
Total_Number_of_Purchases = purchase_data["Purchase ID"].count()
Total_Revenue = purchase_data["Price"].sum()

#create a summary Data Frame
Purchasing_Analysis = pd.DataFrame({"Number Of Unique Items":[Number_of_Unique_Items],
                                   "Average Price":[Average_Purchase_Price],
                                   "Number of Purchases":[Total_Number_of_Purchases],
                                   "Total Revenue":[Total_Revenue]})
#do clean the data and formatting
Purchasing_Analysis["Average Price"] = Purchasing_Analysis["Average Price"].map("${:.2f}".format)
Purchasing_Analysis["Total Revenue"] = Purchasing_Analysis["Total Revenue"].map("${:,.2f}".format)
Purchasing_Analysis = Purchasing_Analysis[["Number Of Unique Items","Average Price","Number of Purchases","Total Revenue"]]

#Display the summary data frame
Purchasing_Analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python

#Establish bins for ages

age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data["Age Group"] = pd.cut(purchase_data["Age"],age_bins, labels=group_names)
Age_Group = purchase_data.groupby("Age Group")
unique_members = Age_Group["SN"].nunique()
Percent_of_players = (unique_members/Total_players)*100

# Create a summary DataFrame

df_AgeDemo = pd.DataFrame({"Percentage of players":Percent_of_players,"Total count":unique_members})

#do a clean formatting 

df_AgeDemo["Percentage of players"] = df_AgeDemo["Percentage of players"].map("{:.2f}".format)
df_AgeDemo = df_AgeDemo[["Total count","Percentage of players"]]
df_AgeDemo 
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
Age_Group = purchase_data.groupby("Age Group")
purchase_count= Age_Group["Age"].count()
avg_purchase_price = Age_Group["Price"].mean()
purchase_total = Age_Group["Price"].sum()
avg_purchase_person = (Age_Group["Price"].sum()/Age_Group["SN"].nunique())

# Create a summary DataFrame
df_Purchasing_Analysis_Age = pd.DataFrame({"Purchase count":purchase_count,"Average Purchase Price":avg_purchase_price,
                           "Total Purchase value":purchase_total,"Ave Total Purchase per person":avg_purchase_person})
#do a clean formatting 

df_Purchasing_Analysis_Age["Average Purchase Price"] = df_Purchasing_Analysis_Age["Average Purchase Price"].map("${:.2f}".format)
df_Purchasing_Analysis_Age["Total Purchase value"]  = df_Purchasing_Analysis_Age["Total Purchase value"].map("${:,.2f}".format)
df_Purchasing_Analysis_Age["Ave Total Purchase per person"] = df_Purchasing_Analysis_Age["Ave Total Purchase per person"].map("${:.2f}".format)
df_Purchasing_Analysis_Age = df_Purchasing_Analysis_Age [["Purchase count","Average Purchase Price","Total Purchase value","Ave Total Purchase per person"]]

df_Purchasing_Analysis_Age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
SN_group = purchase_data.groupby("SN")
Purchase_Count = SN_group["SN"].count()
Average_Purchase_Price = SN_group["Price"].mean()
Total_Purchase_Value = SN_group["Price"].sum()

# Create a summary DataFrame
df_Top_spenders = pd.DataFrame({"Purchase Count":Purchase_Count,"Average Purchase Price":Average_Purchase_Price,
                                 "Total Purchase Value":Total_Purchase_Value})

#Sort the total purchase value column in descending order

df_Top_spenders = df_Top_spenders.sort_values("Total Purchase Value", ascending = False)

#do a clean formatting 

df_Top_spenders["Average Purchase Price"] = df_Top_spenders["Average Purchase Price"].map("${:.2f}".format)
df_Top_spenders["Total Purchase Value"]  = df_Top_spenders["Total Purchase Value"].map("${:,.2f}".format)
df_Top_spenders = df_Top_spenders[["Purchase Count","Average Purchase Price","Total Purchase Value"]]
df_Top_spenders.head()

    
    
      
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Popular_Items = purchase_data.groupby(["Item ID","Item Name"])
purchase_count = Popular_Items["Item ID"].count()
Item_price = Popular_Items["Price"].mean() 
total_purchase_value = Popular_Items["Price"].sum()

# Create a summary DataFrame

df_Popular_Items = pd.DataFrame({"Purchase count":purchase_count,"Item Price":Item_price,
                                   "Total Purchase value":total_purchase_value})

#Sort the purchase count column in descending order
df_Popular_Items = df_Popular_Items.sort_values("Purchase count", ascending = False)



#do a clean formatting 
df_Popular_Items ["Total Purchase value"] = df_Popular_Items ["Total Purchase value"].map("${:,.2f}".format)
df_Popular_Items ["Item Price"] = df_Popular_Items ["Item Price"].map("${:,.2f}".format)
df_Popular_Items = df_Popular_Items[["Purchase count","Item Price","Total Purchase value"]]

df_Popular_Items.head()


    
```
